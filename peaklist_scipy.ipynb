```python
from scipy.signal import find_peaks, normalize
import matplotlib.pyplot as plt
import numpy as np
import json
```

```python
f = open('./test/output/MnO2_Unmilled_Air_xml.json')
res=f.read()
dic = json.loads(res)
```

```python
y = dic["xrdMeasurements"]["xrdMeasurement"]["scan"]["dataPoints"]["intensities"]["#text"]
y = list(map(int, y.split()))
y = np.array(y)
```

```python
step = (79.986 - 10.007)/len(y)
x = np.arange(10.007,79.986,step) 
```

```python
plt.figure(figsize = (9, 5))
plt.plot(x, y)
plt.show()
```

```python
peaks, _ = find_peaks(y)
print("datapoints: %d, peaks: %d." %(len(y), len(peaks)))
```

```python
plt.figure(figsize = (18, 10))
plt.plot(y)
plt.plot(peaks, y[peaks], 'x')
plt.show()
```

```python
a, b = normalize(x, y)
```

```python
plt.figure(figsize = (9, 5))
plt.plot(a, b)
plt.show()
```

```python
mask = np.logical_and(np.logical_and(y>=0, y<80), True)
a = x[mask]
b = y[mask]                      
```

```python
a, b = normalize(a, b)
```

```python
plt.figure(figsize = (9, 5))
plt.plot(a, b)
plt.show()
```
